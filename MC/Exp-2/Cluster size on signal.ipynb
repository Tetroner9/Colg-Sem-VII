```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inferenceration(N, n, i0):
    return np.sqrt(3 * N), 10 * np.log10(np.sqrt(3 * N) ** n /i0)
```

```python
cluster_size = [3, 4, 7, 12, 21]
Q1, SI3, SI4 = [], [], []
```

```python
for N in cluster_size:
    Q, SI = inferenceration(N, 4, 6)
    Q1.append(Q)
    SI4.append(SI)
    Q, SI = inferenceration(N, 3, 6)
    SI3.append(SI)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(cluster_size, Q1, color='red')
plt.stem(cluster_size, Q1)
plt.grid()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 2)
plt.plot(cluster_size, SI4, color='red')
plt.stem(cluster_size, SI4)
plt.plot(cluster_size, SI3, color='red')
plt.stem(cluster_size, SI3)
plt.grid()
```
