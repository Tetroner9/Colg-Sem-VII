```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sc(ta, tc, ca, n):
    m = int(ta / (n * ca))
    k = int(tc / n)
    sc_ans = int(m * k * n)
    return sc_ans
```

```python
print("The System capacity is ", sc(4200, 1001, 12, 7))
```

```python
n1, n2 = [], []

for i in range (1, 5):
    for j in range (1, 5):
        N = np.square(i) + np.square(j) + i * j
        n1.append(N)
        sc_a = sc(4200, 1001, 12, N)
        n2.append(sc_a)
```

```python
print("Value of n1 = ", n1)
print("Value of n2 = ", n2)
```

```python
plt.grid()
plt.stem(n1, n2)
```

```python
n1 = []
n2 = []
N_std = [3, 4, 7, 9, 12, 13, 19, 21, 22]

for i in range (1, 5):
    for j in range (1, 5):
        N = np.square(i) + np.square(j) + i * j
        if N in N_std:
            e = N
            n1.append(e)
            sc_a = sc(4200, 1001, 12, e)
            n2.append(sc_a)
```

```python
print("Value of n1 = ", n1)
print("Value of n2 = ", n2)
```

```python
plt.grid()
plt.stem(n1, n2)
```
