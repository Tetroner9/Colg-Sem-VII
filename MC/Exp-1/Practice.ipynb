```python
def lcm(x, y):
    if x > y:
        greater = x
    else:
        greater = y

    while True:
        if (greater % x == 0) and (greater % y == 0):
            lcm = greater
            break
        greater += 1

    return lcm

lcm (10, 20)
```

```python
def gcd(x, y):
    global gcd
    if x > y:
        smaller = y
    else:
        smaller = x
    for i in range(1, smaller + 1):
        if (x % i == 0) and (y % i == 0):
            gcd = i

    return gcd

gcd(10, 20)
```

```python
def common(a, b):
    c = list(set(a).intersection(b))
    return c
print("Common values of list_1 & list_2 are = " + str(common(list_1, list_2)))
```

```python
x = np.array([1, 4, 6, 4, 2, 9, 10, 12])
y = np.array([3, 7, 9, 3, 2, 7, 6, 10])
x.sort()
y.sort()
```

```python
# Line Plot

plt.plot(x, y, marker = 'x')
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
x1 = x[3]
y1 = y[3]
plt.plot([x1], [y1], 'v', color='red', ms=10)
plt.text(x1, y1, "({:0.0f}, {:0.0f})".format(x1, y1))
plt.show()
```

```python
# Stem Plot

x = np.array([1, 4, 6, 4, 2, 9, 10, 12])
y = np.array([3, 7, 9, 3, 2, 7, 6, 10])

x1 = x[5]
y1 = y[5]
```

```python
plt.stem(x, y)
plt.title('Stem Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.annotate("({:0.0f}, {:0.0f})".format(x1, y1), xy=(x1, y1), xytext=(x1, y1))
plt.show()
```

```python
# Bar Graph
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
plt.title('Bar Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.bar(x, y1, color='orange')
plt.bar(x, y2, bottom=y1, color='g')
plt.show()
```

```python
# Scatter Plot
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
for xy in zip(x, y):
    plt.annotate('(%.0f, %.0f)' % xy, xy=xy)
plt.show()
```
