```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f_d(vm, fc, theta):
    vm = vm * (5 / 18)
    lam = (3 * 10E8) / fc
    fd1 = (vm * np.cos((theta * (np.pi / 180)))) / lam
    return fd1
```

```python
def f_r(fd, fc):
    return fd + fc
```

```python
x = f_d(72, (900 * 10E6), 180)
print(x)
```

```python
theta_array = [0 ,60, 90, 180]
fd, fr = [], []
```

```python
for i in theta_array:
    x = f_d(72, (900 * 10E6), i)
    fd.append(x)
```

```python
for j in fd:
    y = f_r(j, 900E6)
    fr.append(y)
```

```python
print(fd)
print(fr)
```

```python
plt.plot(theta_array, fd)
plt.xlabel('theta')
plt.ylabel('fd')
plt.grid()
plt.show()
```

```python
plt.plot(fr, fd)
plt.xlabel('fr')
plt.ylabel('fd')
plt.grid()
plt.show()
```
