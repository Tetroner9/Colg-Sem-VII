```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pt_dbm(pt):
    return 10 * np.log10(pt)
```

```python
def log_distance(ref_dis_loss, n, d, do):
    return ref_dis_loss + 10 * n * np.log10(d / do)
```

```python
def log_normal(pl1, shadow_eff):
    return pl1 + shadow_eff
```

```python
def receiver_pow(pt, pl):
    return pt - pl
```

```python
pt = pt_dbm(4 * 1000)
print(pt)
pl1 = log_distance(-32, 4, 3000, 100)
print(pl1)
pl2 = log_normal(pl1, 10.5)
print(pl2)
```

```python
pr1 = receiver_pow(pt, pl1)
print(pr1)
pr2 = receiver_pow(pt, pl2)
print(pr2)
```

```python
PR1 = []
PR2 = []
N = []
for i in range(1, 11):
    N.append(i)
    pr1 = pt - log_distance(-32, i, 3000, 100)
    PR1.append(pr1)
    pr2 = pt - log_normal(log_distance(-32, i, 3000, 100), 10.5)
    PR2.append(pr2)
```

```python
plt.plot(N, PR1, label='PR1')
plt.plot(N, PR2, label='PR2')
plt.xlabel('N')
plt.ylabel('PR')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rx1 = [pt - log_distance(-32, i, 3000, 100) for i in range (1, 11)]
rx2 = [pt - log_normal(log_distance(-32, i, 3000, 100), 10.5) for i in range (1, 11)]
```

```python
plt.plot(list(range(1, 11)), rx1, label='rx1')
plt.plot(list(range(1, 11)), rx2, label='rx2')
plt.xlabel('N')
plt.ylabel('Rx')
plt.legend()
plt.grid(True)
plt.show()
```
